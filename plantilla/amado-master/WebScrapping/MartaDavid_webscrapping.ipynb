```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json as json
```

```python
# Inicia el WebDriver directamente (sin necesidad de especificar la ruta si el controlador está en PATH)
driver = webdriver.Chrome()

# Abre la página principal de Google
driver.get("https://www.google.com/")

def get_card_links(base_url, max_pages):
    all_links = {}

    for page in range(0, max_pages + 1):
        url = f"{base_url}&site={page}"
        driver.get(url)
        
        try:
            # Espera a que los elementos de las cartas estén presentes en la página
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'galleryBox')))
            
            # Encuentra todos los elementos de enlace de cartas
            cards = driver.find_elements(By.CSS_SELECTOR, 'a.card.text-center.w-100.galleryBox')
            
            for card in cards:
                img_element = card.find_element(By.CSS_SELECTOR, 'img.card-img-top')  # Encuentra la imagen dentro de la carta
                img_url = img_element.get_attribute('src')  # Obtiene la URL de la imagen
                name_card = card.find_element(By.CSS_SELECTOR, 'h2.card-title').text.strip()  # Extrae el nombre de la carta
                
                if img_url:
                    all_links[name_card] = img_url
        except Exception as e:
            print(f"Error al procesar la página {page}: {e}")
        
        time.sleep(2)  # Para evitar bloqueos por hacer muchas solicitudes rápidamente

    return all_links


# URL base sin el parámetro de página
base_url = 'https://www.cardmarket.com/es/Magic/Products/Singles?mode=gallery'

# Número máximo de páginas a recorrer
max_pages = 15

# Ejecuta la función y obtiene los enlaces
card_links = get_card_links(base_url, max_pages)

# Imprime los enlaces obtenidos
for link in card_links:
    print(link)

# Cierra el navegador
driver.quit()

```

```python
output_file = "card_list.json"
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump({"cards": card_links}, json_file, ensure_ascii=False, indent=4)

print(f"Enlaces guardados en {output_file}.")
```
