```python
import requests

# Lista de Pokémon (asumiendo que la tienes declarada)
# pokemones = ['Bulbasaur', 'Ivysaur', 'Venusaur', 'Charmander', 'Charmeleon', ...]

# Diccionario que mapea el nombre de la estadística a su ID en la base de datos
stat_names = {
    'hp': 1,
    'attack': 2,
    'defense': 3,
    'special-attack': 4,
    'special-defense': 5,
    'speed': 6
}

# Función para obtener las estadísticas de un Pokémon por su número de Pokédex
def get_pokemon_stats(pokemon_id):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_id}'
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        stats = {stat['stat']['name']: stat['base_stat'] for stat in data['stats']}
        return stats
    else:
        print(f"Error al obtener datos para el Pokémon con ID {pokemon_id}")
        return None

# Generar y guardar los inserts SQL para cada Pokémon, uno a uno
def generate_and_save_inserts(pokemones, filename='stats.sql'):
    with open(filename, 'w') as file:
        for pokemon_id in range(1, 1026):  # Número de Pokédex del 1 al 1025
            if pokemon_id <= len(pokemones):  # Asegurarse de que la lista tiene suficientes Pokémon
                pokemon_name = pokemones[pokemon_id - 1]  # Nombre del Pokémon según su posición en la lista
                stats = get_pokemon_stats(pokemon_id)
                
                if stats:
                    insert_statements = []
                    for stat_name, stat_value in stats.items():
                        stat_id = stat_names.get(stat_name)
                        if stat_id:
                            insert_statements.append(f"((SELECT id FROM pokemones WHERE name = '{pokemon_name}'), {stat_id}, {stat_value})")
                    
                    insert_statement = f"-- Inserción para {pokemon_name}\n" + \
                                       f"INSERT INTO stat_values (pokemon_id, stat_id, stat_value) VALUES\n" + \
                                       ",\n".join(insert_statements) + ";"
                    file.write(insert_statement + "\n\n")
                    # print(f"Insert para {pokemon_name} (ID: {pokemon_id}) guardado correctamente.")
                else:
                    print(f"Error al procesar {pokemon_name} (ID: {pokemon_id}), no se guardó en el archivo.")
            else:
                print(f"ID {pokemon_id} no tiene un Pokémon en la lista proporcionada.")

# Llamar a la función para generar y guardar los inserts
generate_and_save_inserts(pokemones)
print("Los inserts han sido procesados y guardados en 'stats.sql'.")

```
